```python
import pandas as pd
import numpy as np

df = pd.DataFrame({'col1': ['A', 'A', 'B', np.nan, 'D', 'C'],
                   'col2': [2, 1, 9, 8, 7, 4],
                   'col3': [0, 1, 9, 4, 2, 3]})
print(df)
print(df.sort_values(by='col1'))  # 按 col1 升序，NaN 置後
print(df.sort_values(by=['col1', 'col2'], ascending=[False, True]))  # 多欄位混合排序
```
