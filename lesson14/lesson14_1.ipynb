```python
import matplotlib as mpl
from matplotlib.font_manager import fontManager

fontManager.addfont("ChineseFont.ttf")
mpl.rc('font', family="ChineseFont")
```

```python
import matplotlib.pyplot as plt

student1 = {"國文":78, "英文":80, "數學":92, "歷史":75, "探索":85}
student2 = {"國文":92, "英文":60, "數學":85, "歷史":62, "探索":73}

subjects = list(student1.keys())
scores1 = list(student1.values())
scores2 = list(student2.values())

figure = plt.figure(dpi=100, figsize=(8,5), facecolor='lightgray')
axes = figure.add_subplot(1,1,1)
axes.set_title("學生分數比較")
axes.set_xlabel("科目")
axes.set_ylabel("分數")

# 在 plot 時加上 label，然後呼叫 legend()
# 自訂line的style

axes.plot(subjects, scores1, marker='o', label='學生1', linestyle='--')
axes.plot(subjects, scores2, marker='s', label='學生2', linestyle='-.')
axes.legend()

# 圖表左右和上下留白,避免貼齊邊框,由程式控制距離
# 請使用改變 xticks 和 yticks 的方式來達成
# 使用最簡單的方式設定左右多一些留白
axes.set_xlim(-0.5, len(subjects)-0.5)

# 請改變 yticks 的刻度範圍和間距,起始值是我的最小分數,終止值是我的最大分數
axes.set_yticks(range(min(min(scores1), min(scores2)) - 10 , max(max(scores1), max(scores2)) + 10, 10))


plt.show()
```
