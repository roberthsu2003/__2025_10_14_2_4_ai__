```python
import matplotlib.pyplot as plt

# 1.製作figure
fig = plt.figure()

# 2.在1.的figure上製作一個axes
ax_1 = fig.add_subplot(1, 1, 1)

# 3.將圖的數據儲存在axes中
X = [0, 1, 2]
Y = [3, 4, 5]

ax_1.plot(X, Y,color='blue', marker='s', linestyle='solid',linewidth=1, markersize=8)

# 4.顯示圖
plt.show()
```

```python
import matplotlib.pyplot as plt
# 1.製作figure
fig = plt.figure()

# 2.1*3的矩陣圖表的第1個axes
ax_1 = fig.add_subplot(1, 3, 1)

# 設定ax_1的標題
ax_1.set_title(label='Ax 1')

# 2.1*3的矩陣圖表的第2個axes
ax_2 = fig.add_subplot(1, 3, 2)
# 設定ax_2的標題
ax_2.set_title(label='Ax 2')

# 2.1*3的矩陣圖表的第3個axes
ax_3 = fig.add_subplot(1, 3, 3)
# 設定ax_3的標題
ax_3.set_title(label='Ax 3')

# 3.顯示圖
plt.show()
```

```python
import matplotlib.pyplot as plt
# 1.製作figure
fig_1 = plt.figure()
# 2.在1.的figure上製作一個axes
ax_1 = fig_1.add_subplot(1, 1, 1)
# 3.將圖的數據儲存在axes中
X = [0, 1, 2]
Y = [3, 4, 5]
ax_1.plot(X, Y)

fig_2 = plt.figure()
ax_2 = fig_2.add_subplot(1, 1, 1)
X = [2, 3, 6]
Y = [2, 1, 2]
ax_2.plot(X, Y)

# 4.顯示圖
plt.show()
```
