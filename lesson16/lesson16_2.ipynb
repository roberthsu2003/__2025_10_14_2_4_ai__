```python
import pandas as pd

series1 = pd.Series([4.5, 7.2, -5.3, 3.6],index=['d','b','a','c'])
series1
```

```python
# Series的實體方法reindex()
series1.reindex(['a','b','c','d'])
```

```python
series1.reindex(['a','b','c'])
```

```python
import numpy as np

frame = pd.DataFrame(
                    data=np.arange(9).reshape((3,3)),
                    index = ['a','c','d'],
                    columns = ['台北','台中','高雄']
                    )
frame
```

```python
frame.reindex(columns=['台北','台中'],index=['a','c'])
```

```python
# 實體方法drop()
frame1 = pd.DataFrame(np.arange(16).reshape(4,4),
                   index=['台北','台中', '台南', '高雄'],
                   columns=['one', 'two', 'three', 'four'])
frame1
```

```python
frame1.drop(index=['台北','台中'],columns=['one','two'])
```
