```python
import requests
from requests import JSONDecodeError

try:
    response = requests.get("http://example.com")
    print(response.status_code)  # 狀態碼
    print(response.text)         # 獲取網頁內容 (文字)
    print(response.json())       # 若回應為JSON格式，可直接解析為dict
except JSONDecodeError:
    print("內容不是json的格式")
except Exception:
    print("發生錯誤")
else:
    print("沒有發生錯誤")
```
