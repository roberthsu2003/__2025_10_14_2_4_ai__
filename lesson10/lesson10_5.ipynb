```python
class Person:
    def __init__(self,name:str,age:int): #自訂的初始化
        self.name = name
        self.age = age

    #實體方法
    def description(self):
        print(f"我的名字叫{self.name}")
        print(f"我的age是:{self.age}")
```

```python
p1 = Person(name="Robert", age=40)
p1.description()
```

```python
# Student繼承於Person
# Person是父類別
# Student是子類別

class Student(Person):
    def __init__(self,name:str,age:int,chinese:int=60,english:int=60,math:int=60):
        super().__init__(name,age)
        self.chinese = chinese
        self.english = english
        self.math = math
    
```

```python
s1 = Student(name="robert",age=40,english=86,math=92)
print(s1.chinese)
print(s1.english)
print(s1.math)
```
