```python
class Person:
    def __init__(self,name:str,age:int): #自訂的初始化
        self.name = name
        self.age = age

    #實體方法
    def description(self):
        print(f"我的名字叫{self.name}")
        print(f"我的age是:{self.age}")
```

```python
p1 = Person(name="Robert", age=40)
p1.description()
```
