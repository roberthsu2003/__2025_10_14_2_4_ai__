```python
import matplotlib as mpl
from matplotlib.font_manager import fontManager

fontManager.addfont("ChineseFont.ttf")
mpl.rc('font', family="ChineseFont")
```

```python
import matplotlib.pyplot as plt

student1 = {"":78, "":80, "":92, "":75, "":85}
student2 = {"":92, "":60, "":85, "":62, "":73}

subjects = list(student1.keys())
scores1 = list(student1.values())
scores2 = list(student2.values())

figure = plt.figure(dpi=100, figsize=(8,5), facecolor='lightgray')
axes = figure.add_subplot(1,1,1)
axes.set_title("")
axes.set_xlabel("")
axes.set_ylabel("")

#  plot  label legend()
# linestyle

axes.plot(subjects, scores1, marker='o', label='1', linestyle='--')
axes.plot(subjects, scores2, marker='s', label='2', linestyle='-.')
axes.legend()

plt.show()
```
