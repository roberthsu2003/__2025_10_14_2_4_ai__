```python
class Person:
    def __init__(self,name:str,age:int): #自訂的初始化
        self.name = name
        self.age = age

    #實體方法
    def description(self):
        print(f"我的名字叫{self.name}")
        print(f"我的age是:{self.age}")
```

```python
p1 = Person(name="Robert", age=40)
p1.description()
```

```python
# Student繼承於Person
# Person是父類別
# Student是子類別

class Student(Person):
    @classmethod
    def nickname(cls):
        print("我是class的method")

    @staticmethod
    def heroname():
        print("我是static的method")
    
    def __init__(self,name:str,age:int,chinese:int=60,english:int=60,math:int=60):
        super().__init__(name,age)
        self.__chinese = chinese
        self.__english = english
        self.__math = math

    @property
    def chinese(self)->int:
        return self.__chinese
    
    @property
    def english(self)->int:
        return self.__english
    
    @property
    def math(self)->int:
        return self.__math
    
    #覆寫(override)
    def description(self):
        super().description()
        print(f"國文:{self.chinese}")
        print(f"英文:{self.english}")
        print(f"數學:{self.math}")
    
```

```python
s1 = Student(name="robert",age=40,english=86,math=92)
s1.description()
```

```python
print(s1.chinese)
s1.description()
```
