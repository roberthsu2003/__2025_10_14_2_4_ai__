```python
import pandas as pd
data = {'Taipei':35000, 'Taoyuan': 71000, 'Taichung': 16000, 'Kaohsiung':5000}
s1 = pd.Series(data)
type(s1)
s1
```

```python
import matplotlib.pyplot as plt

index = s1.index
values = s1.values
figure = plt.figure()
axes = figure.add_subplot()
axes.bar(index, values)
plt.show()
```
